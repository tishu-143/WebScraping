```


from bs4 import BeautifulSoup
import requests
import pandas as pd

web_url = "https://boston.craigslist.org/search/sof"

jobs_list = {}
job_no = 0
while True:
  response = requests.get(web_url)
  data = response.text

  soup = BeautifulSoup(data,'html.parser')

  jobs = soup.find_all('p',{'class':'result-info'})

  for job in jobs:
    title = job.find('a',{'class':'result-title'}).text
    add = job.find('span',{'class':'result-hood'})
    address = add.text if add else "N/A"
    link = job.find('a',{'class':'result-title'}).get('href')
    job_response = requests.get(link)
    job_data = job_response.text
    job_soup = BeautifulSoup(job_data,'html.parser')
    job_att = job_soup.find('p',{'class':'attrgroup'})
    job_attribute = job_att.text if job_att else "N/A"
    job_description = job_soup.find('section',{'id':'postingbody'}).text
    job_no+=1
    jobs_list[job_no] = [title,address,link,job_attribute,job_description]

    print("Job Title = {}\nJob Address = {}\nJob link = {}\nJob Attribute = {}\nJob Description = {}------------------\n\n".format(title,address,link,job_attribute,job_description))
  
  url_tag = soup.find('a',{'title':'next page'})

  if url_tag.get('href'):
    web_url = 'https://boston.craigslist.org'+url_tag.get('href')
  else:
    break
npo_job_df = pd.DataFrame.from_dict(jobs_list, orient = 'index', columns = ['Job Title', 'Location', 'Job Link', 'Job Attribute', 'Job Description'])
print("Total Jobs = {}".format(job_no))
npo_job_df.head()
npo_job_df.to_csv('/content/drive/My Drive/DataSets/Udemy_WebScraping/Craigslist_Jobs')
```

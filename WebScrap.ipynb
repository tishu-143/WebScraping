```
from bs4 import BeautifulSoup
import requests
```

## **Extracting response from url**

```
web_url = "https://boston.craigslist.org/search/sof"

response = requests.get(web_url)

response
```

# **Printing the html data**

```
data = response.text

data
```

# **Creating beautifulsoup object**

```
soup = BeautifulSoup(data,"html.parser")

tags = soup.find_all('a')
```

# **Browsing through the anchor tags**

```
for tag in tags:
  print(tag.get('href'))
```

# **Scraping the titles of jobs in craigslist**

```
titles = soup.find_all('a',{'class':'result-title'})

for title in titles:
  print(title.text)
```

```
addresses = soup.find_all('span',{'class':'result-hood'})

for address in addresses:
  print(address.text)
```

# **Scraping all information at once for a job**

```
jobs = soup.find_all('p',{'class':'result-info'})

for job in jobs:
  title = job.find('a',{'class':'result-title hdrlnk'}).text
  address = job.find('span',{'class':'result-hood'})
  add = address.text[2:-1] if address else "N/A"
  date = job.find('time',{'class':'result-date'}).text
  link = job.find('a',{'class':'result-title'}).get('href')

  job_response = requests.get(link)
  job_data = job_response.text
  job_soup = BeautifulSoup(job_data,'html.parser')
  job_description = job_soup.find('section',{'id':'postingbody'}).text
  job_att_tag = job_soup.find('p',{'class':'attrgroup'})
  job_att = job_att_tag.text if job_att_tag else "N/A"
  print("Title = {}, \nAddress = {}, \ndate = {}, \nlink = {}, \nJob Attribute = {}, \nJob Description = {}\n--------------------------------------------\n".format(title,add,date,link, job_att,job_description))
```
